# DATA CLEANING AND PREPOCESSING

```python
 df.shape
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
 df.dtypes
```

```python
df['sex'].value_counts()
```

## level Encoding

```python
df['sex']= df['sex'].map({'F':0 , 'M':1})
```

```python
df['address'].value_counts()
```

```python
df['address']=df['address'].map({'U':0 , 'R':1})
```

```python
df.head(8)
```

```python
df.rename(columns = {
    'sex': 'is_male',
    'address':'is_rular_address'
}, inplace = True)
```

```python
df['is_rular_address']
```

## hot encoding

```python
df = pd.get_dummies(df, columns = ['Mjob','Fjob'], drop_first = True)
```

```python
df.head()
```

```python
df['famsize'].value_counts()
```

```python
df['famsize'] = df['famsize'].map({'LE3':1,'GT3':0})
```

```python
df.rename(columns = {
    'famsize':'is_famsize_LE3'
}, inplace = True)
```

```python
df['is_famsize_LE3']
```

```python
df = pd.get_dummies(df , columns  = ['school'])
```

```python
df['Pstatus'].value_counts()
```

```python
df['Pstatus'] = df['Pstatus'].map({'T':1 , 'A':0})
df.rename(columns = {
    'Pstatus': 'is_parents_living_together'
}, inplace = True)
```

```python
df['reason'].value_counts()
```

```python
df = pd.get_dummies(df , columns = ['reason','guardian'] , drop_first = True)
```

```python
df['schoolsup'].value_counts()
```

```python

df['famsup'] = df['schoolsup'].map({'no':0 , 'yes':1})
df['paid'] = df['paid'].map({'no':0 , 'yes':1})
df['activities'] = df['activities'].map({'no':0 , 'yes':1})
df['nursery'] = df['nursery'].map({'no':0 , 'yes':1})
df['higher'] = df['higher'].map({'no':0 , 'yes':1})
df['internet'] = df['internet'].map({'no':0 , 'yes':1})
df['romantic'] = df['romantic'].map({'no':0 , 'yes':1})
df['schoolsup'] = df['schoolsup'].map({'no':0 , 'yes':1})
```

```python
df.rename(columns = {
    'schoolsup':'is_sch_supp_available',
    'famsup':'is_fam_supp_available',
    'paid':'is_extra_paid_classes',
    'activities':'extra_curricular_activities'
},inplace = True)
```

```python
df['romantic'].value_counts()
```

```python
df = df.astype(int)
```

```python
df.dtypes
```

# Feature Engineering and  Extraction


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop(['G1','G2','G3'],axis = 1)
y = df['G1']
X_train , X_test , y_train , y_test = train_test_split(X , y, test_size = 0.20, random_state= 42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test , y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1 - r2)  * (n-1) / (n-p-1))
```
